Nettoyage des colonnes Points_forts, Expérience, Compétence

```python
import re
import string

def nettoyer_texte(texte):
    if pd.isna(texte):
        return ""
    texte = re.sub(r'[\r\n\t]+', ' ', texte)
    texte = texte.translate(str.maketrans('', '', string.punctuation))
    texte = re.sub(r'\s+', ' ', texte)
    return texte.strip()

colonnes_a_nettoyer = ['Points_forts', 'Expérience', 'Compétence']

for col in colonnes_a_nettoyer:
    df_profile_cleaned[col] = df_profile_cleaned[col].apply(nettoyer_texte)
```

```python
df_profile_cleaned
```

suppresion des profils qui ont des titres exclu de l'IT

```python
mots_cles_exclus = [
    "ressources humaines", "rh", "recrutement", "recruteur",
    "marketing", "communication", "finance", "comptabilité",
    "juridique", "assistant", "assistante", "commercial", "vente",
    "banque", "audit", "contrôle de gestion", "gestionnaire", "achats", "relation client"
]

def est_exclu(profil):
    if pd.isna(profil):
        return False
    profil_lower = profil.lower()
    return any(mot in profil_lower for mot in mots_cles_exclus)

df_profile_cleaned = df_profile_cleaned[~df_profile_cleaned['Profil'].apply(est_exclu)].copy()
```

```python
df_profile_cleaned
```

extraction des mots clé de contrat de la colonne Profil 

```python
import re

contrats_clé = [
    "CDI", "CDD", "Alternance", "Apprentissage",
    "contrat d'apprentissage", "contrat de professionnalisation"
]

def extraire_contrat(texte):   
    if pd.isna(texte):
        return ""
    contrats_trouvé = [c for c in contrats_clé if re.search(rf'\b{re.escape(c)}\b', texte, flags=re.IGNORECASE)]
    return ", ".join(set(contrats_trouvé))

df_profile_cleaned['Contrat'] = df_profile_cleaned['Profil'].apply(extraire_contrat)
```

Regroupement des métiers

```python
def regrouper_metier(profil):
    if pd.isna(profil):
        return "Inconnu"

    profil_lower = profil.lower()

    if "data analyst" in profil_lower:
        return "Data Analyst"
    elif "data engineer" in profil_lower:
        return "Data Engineer"
    elif "data scientist" in profil_lower:
        return "Data Scientist"
    elif "développeur" in profil_lower or "developer" in profil_lower:
        return "Développeur"
    elif "ingénieur logiciel" in profil_lower or "software engineer" in profil_lower:
        return "Ingénieur Logiciel"
    elif "admin" in profil_lower and ("système" in profil_lower or "réseau" in profil_lower):
        return "Administrateur Systèmes/Réseaux"
    elif "devops" in profil_lower:
        return "DevOps"
    elif "consultant bi" in profil_lower or "business intelligence" in profil_lower:
        return "Consultant BI"
    elif "architecte" in profil_lower:
        return "Architecte SI"
    elif "chef de projet" in profil_lower or "scrum master" in profil_lower:
        return "Chef de projet/Scrum Master"
    elif "cybersécurité" in profil_lower or "sécurité informatique" in profil_lower:
        return "Cybersécurité"
    else:
        return "Autre"

df_profile_cleaned["Metier_regroupe"] = df_profile_cleaned["Profil"].apply(regrouper_metier)
```
